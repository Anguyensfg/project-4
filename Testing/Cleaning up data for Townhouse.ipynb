```python
#Dependencies
import pandas as pd
import sqlite3
```

```python
#Read 1st csv files as pandas dataframe and view
csv_file = "Resources/Property_Sale.csv"
PS_df = pd.read_csv(csv_file)
PS_df.head()
```

```python
#Rename columns of 1st dataframe
PS_final_df = PS_df.rename(columns={"lat": "latitude", "lon": "longitude","loc_pid":"postcode"})

#display
PS_final_df.head()
```

```python
# Dropping the rows with "NaN"
# identify partial string
PS_final_df = PS_final_df.dropna()

#display
PS_final_df
```

```python
# Filtering out ACT state"
df = PS_final_df[(PS_df['state'] == 'ACT')]
df
```

```python
# Filtering out ACT state and House"
df = PS_final_df[(PS_df['property_type'] == 'townhouse') & (PS_df['state'] == 'ACT')]
df
```

```python
# Create new dataframe
newdf =df.copy()
newdf
```

```python
# Modify Postcode
newdf['postcode'] = newdf['postcode'].str[3:]
newdf
```

```python
# Crop section of dataframe
part1_df = newdf.iloc[:,7:8]
part1_df
```

```python
# Crop section of dataframe
part2_df = newdf.iloc[:,1:2]
part2_df
```

```python
# Crop section of dataframe
part3_df = newdf.iloc[:,9:10]
part3_df
```

```python
# Combined crop sections
final_combined_df= pd.concat([part2_df, part1_df, part3_df], axis=1)
final_combined_df
```

```python
# Create output CSV file
final_combined_df.to_csv('Resources\ACT_Townhouses.csv')
```
